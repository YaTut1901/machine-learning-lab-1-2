```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

```python
# 1) Generate synthetic linear data
x, y = make_regression(
    n_samples=150, n_features=1, noise=15.0, bias=5.0, random_state=42
)
```

```python
# 2) Fit a simple linear regression
model = LinearRegression().fit(x, y)
y_pred = model.predict(x)

print(f"Intercept: {model.intercept_:.2f}, Coef: {model.coef_[0]:.2f}")
```

```python
# 3) Plot data points and the regression line
order = np.argsort(x[:, 0])
plt.figure(figsize=(7, 5))
plt.scatter(x[:, 0], y, color="steelblue", alpha=0.75, label="Data")
plt.plot(x[order, 0], y_pred[order], color="crimson", linewidth=2, label="Regression line")
plt.title("Simple Linear Regression (synthetic)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
